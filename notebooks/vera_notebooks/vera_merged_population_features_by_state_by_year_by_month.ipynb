```python
import pandas as pd
import numpy as np
```

```python
# the features by state season csv
features_df = pd.read_csv('../../raw_data/features_by_state_year_month.csv', low_memory=False)
features_df.drop(columns= ['state_population', 'state_military_bases', 'state_cities'], inplace = True)
# the population by state season csv
population_df = pd.read_csv('../../raw_data/population_by_state_year_month.csv', low_memory=False)
```

```python
features_df = features_df[features_df['year_number']>1974]
features_df
```

```python
population_df
```

```python
# merge the 2 dataframes
merged_df = features_df.merge(population_df, how='outer', left_on=["state", 'year_number', 'month_number'], right_on=["state", 'year_number', 'month_number'])
```

```python
merged_df.dropna(axis=0, how='any', inplace = True)
```

```python
merged_df
```

```python
merged_df.to_csv('../../raw_data/merged_df_by_year_by_month.csv',header=True,index=False)
```
